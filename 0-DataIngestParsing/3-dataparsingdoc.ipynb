```python
from langchain_community.document_loaders import Docx2txtLoader,UnstructuredWordDocumentLoader
```

```python
print("docx2txtloader")
try:
    docx_loader=Docx2txtLoader("data/docx/langdocx.docx")
    docx=docx_loader.load()
    print(len(docx))
except Exception as e:
    print(f"error in parsing data :{e}")
```

```python
print("unstructuredworddocumentloader")
try:
    load_docx=UnstructuredWordDocumentLoader("data/docx/langdocx.docx",mode="elements")
    doc=load_docx.load()
    print(len(doc))
except Exception as e:
    print(f"eroor {e}")
```
